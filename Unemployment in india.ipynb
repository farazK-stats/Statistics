```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Unemployment in india.csv')
```

```python
df.head()
```

```python
# Check missing values
df.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.isnull().sum()
```

```python
# Data columns have some extra spaces. Removing them
df.columns = df.columns.str.strip()
```

## Convert the 'Date' column to datetime format

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

## Visualize the Unemployment Rate Over Time

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=df["Date"], y=df["Estimated Unemployment Rate (%)"], marker="o", color="blue")

plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.grid()

plt.show()
```

**Interpretation:**     
From June 2019 to April 2020, the unemployment rate remained stable between 8% and 10%, indicating a steady job market. However, in May and June 2020, it spiked sharply to 24% and 25%, respectively, due to the impact of COVID-19 lockdowns, which led to widespread job losses and economic slowdown. By July 2020, the unemployment rate dropped to 12.5%, signaling the beginning of economic recovery as restrictions were eased and businesses started reopening. This trend highlights the severe impact of the pandemic on employment, followed by a gradual rebound.

## Compare Unemployment in Different Regions

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Region", y="Estimated Unemployment Rate (%)", data=df)

plt.title("Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=90)

plt.show()
```

**Interpretation:**        
The box plot illustrates the variation in unemployment rates across different regions, highlighting significant disparities. Some states, such as Haryana, Rajasthan, and Tripura, exhibit consistently high unemployment, with extreme outliers reaching above 70%, indicating severe job market instability. In contrast, regions like Meghalaya, Gujarat, and Karnataka maintain lower and more stable unemployment rates, suggesting stronger economic conditions. The presence of multiple outliers in various states suggests that certain regions experienced temporary but severe unemployment spikes, likely due to economic disruptions such as the COVID-19 pandemic. This analysis underscores the uneven distribution of unemployment, emphasizing the need for targeted job creation policies in high-unemployment areas to ensure economic stability.

## Analyze Employment Trends

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=df["Date"], y=df["Estimated Employed"], marker="o", color="green")

plt.title("Employment Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Employed People")
plt.xticks(rotation=45)
plt.grid()

plt.show()
```

**Interpretation:**    
From July 2019 to April 2020, the number of employed people remained relatively stable between 0.7 and 0.8, indicating a steady job market. However, in May 2020, employment dropped sharply to 0.54, reflecting a major job loss, likely due to COVID-19 lockdowns that forced businesses to shut down and disrupted economic activities. In June 2020, employment slightly recovered to 0.6, showing initial signs of economic adjustment. By July 2020, employment rose further to 0.74, indicating a strong rebound as restrictions eased, businesses reopened, and people returned to work. This trend highlights the severe impact of the pandemic on employment, followed by a gradual recovery as economic activities resumed.

## Analyze Labour Participation Rate

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=df["Date"], y=df["Estimated Labour Participation Rate (%)"], marker="o", color="red")

plt.title("Labour Participation Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Labour Participation Rate (%)")
plt.xticks(rotation=45)
plt.grid()

plt.show()
```

**Interpretation:**        
From July 2019 to March 2020, the labour rarticipation rate remained fairly stable between 44% and 43%, suggesting a consistent labor market. However, in April and May 2020, the labour participation rate dropped sharply, reaching its lowest point around 35% in May 2020. This indicates that a large portion of the workforce exited the labor market, likely due to COVID-19 lockdowns, job losses, and economic uncertainty. After this sharp decline, the labour participation rate started recovering in June and July 2020, reaching around 41% by July 2020. This suggests that as restrictions eased and businesses reopened, more people re-entered the workforce, either by returning to their jobs or actively seeking new employment.
