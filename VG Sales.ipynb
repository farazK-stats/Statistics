## Total Global sales by region

```python
region_sales = {
    'Region': ['North America', 'Europe and Africa', 'Japan', 'Rest of the World', 'Global'],
    'Global Sales (in million)': [
        df['na_sales'].sum(),
        df['pal_sales'].sum(),
        df['jp_sales'].sum(),
        df['other_sales'].sum(),
        df['total_sales'].sum()
    ]
}

# Convert the dictionary to a DataFrame
region_sales_df = pd.DataFrame(region_sales)

# Show the table
region_sales_df
```

## Analyze the Relationship Between Critic Score and Global Sales

```python
# Plot a scatter plot to visualize the relationship between critic score and global sales
import matplotlib.pyplot as plt
import seaborn as sns

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.regplot(x='critic_score', y='total_sales', data=df, scatter_kws={'s': 50}, line_kws={"color": "red"})

# Add labels and title
plt.title('Relationship Between Critic Score and Global Sales')
plt.xlabel('Critic Score')
plt.ylabel('Global Sales (in millions)')

# Show the plot
plt.show()
```

Interpretation: The dots are often close to the line and the line goes upward, which means that higher critic scores usually lead to higher global sales

## Top 5 publishers by specific regions and global sales

```python
# Group by publisher and sum the sales for each region (NA, EU, JP, Other) and global sales
# ascending=False sorts the data in descending order (from highest to lowest).
# ascending=True sorts the data in ascending order (from lowest to highest).

na_sales_by_publisher = df.groupby('publisher')['na_sales'].sum().sort_values(ascending=True).head(5)
pal_sales_by_publisher = df.groupby('publisher')['pal_sales'].sum().sort_values(ascending=True).head(5)
jp_sales_by_publisher = df.groupby('publisher')['jp_sales'].sum().sort_values(ascending=True).head(5)
other_sales_by_publisher = df.groupby('publisher')['other_sales'].sum().sort_values(ascending=True).head(5)
global_sales_by_publisher = df.groupby('publisher')['total_sales'].sum().sort_values(ascending=True).head(5)

# Create subplots for the top 5 publishers by region and global sales
fig, axes = plt.subplots(1, 5, figsize=(50, 12))



# Plot Top 5 Publishers by NA Sales (Horizontal)
# kind='bar' (which is vertical)  kind='barh' (which is horizontal)
na_sales_by_publisher.plot(kind='barh', ax=axes[0], color='skyblue')
axes[0].set_title('Top 5 Publishers by North America Sales')
axes[0].set_xlabel('NA Sales (in millions)')
axes[0].set_ylabel('Publisher')

# Plot Top 5 Publishers by EU Sales (Horizontal)
pal_sales_by_publisher.plot(kind='barh', ax=axes[1], color='green')
axes[1].set_title('Top 5 Publishers by European Sales')
axes[1].set_xlabel('Europe and Africa Sales (in millions)')
axes[1].set_ylabel('Publisher')

# Plot Top 5 Publishers by JP Sales (Horizontal)
jp_sales_by_publisher.plot(kind='barh', ax=axes[2], color='black')
axes[2].set_title('Top 5 Publishers by Japan Sales')
axes[2].set_xlabel('JP Sales (in millions)')
axes[2].set_ylabel('Publisher')

# Plot Top 5 Publishers by Other (Rest of the World) Sales (Horizontal)
other_sales_by_publisher.plot(kind='barh', ax=axes[3], color='yellow')
axes[3].set_title('Top 5 Publishers by Rest of the World Sales')
axes[3].set_xlabel('Other Sales (in millions)')
axes[3].set_ylabel('Publisher')

# Plot Top 5 Publishers by Global Sales (Horizontal)
global_sales_by_publisher.plot(kind='barh', ax=axes[4], color='salmon')
axes[4].set_title('Top 5 Publishers by Global Sales')
axes[4].set_xlabel('Global Sales (in millions)')
axes[4].set_ylabel('Publisher')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Genre-Wise Total Sales in Each Region

```python
# Group by genre and sum sales for each region
genre_sales = df.groupby('genre')[['na_sales', 'pal_sales', 'jp_sales', 'other_sales']].sum()

# Plotting genre-wise sales for each region
genre_sales.plot(kind='barh', figsize=(12, 8))
plt.title('Genre-Wise Total Sales by Region', fontsize=15)
plt.xlabel('Genre')
plt.ylabel('Total Sales (in millions)')
plt.xticks(rotation=45)  # Rotate genre labels for readability
plt.legend(title='Region')  # Add a legend for regions
plt.tight_layout()  # Adjust layout to fit all labels
plt.show()
```

## Platform-Wise Global Sales Analysis

```python
# Group by platform and calculate total global sales
platform_sales = df.groupby('console')['total_sales'].sum()

# Sort platforms by sales in descending order and take the top N contributors
top_n = 6  # You can change this value to any number, e.g., 5, 10, etc.
top_platforms = platform_sales.sort_values(ascending=False).head(top_n)

# Calculate the percentage contribution of each platform to global sales
total_sales = platform_sales.sum()
platform_percentage = (top_platforms / total_sales) * 100

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(platform_percentage, labels=platform_percentage.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Title and other chart settings
plt.title(f'Top {top_n} Platforms by Global Sales Percentage', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular

# Show the chart
plt.show()
```

## Top Publishers by sales across all regions

```python
import matplotlib.pyplot as plt

# Define the regions to analyze
regions = ['na_sales', 'pal_sales', 'jp_sales', 'other_sales', 'total_sales']  # Total sales for global

# Loop through each region and create a bar chart
for region in regions:
    # Group by publisher and sum sales for the selected region
    publisher_sales_region = df.groupby('publisher')[region].sum()

    # Plot the top 5 publishers for the selected region
    plt.figure(figsize=(10, 5))  # Set figure size for each plot
    publisher_sales_region.sort_values(ascending=False).head(5).plot(kind='bar', color='skyblue')
    plt.title(f'Top 5 Publishers by {region.replace("_", " ").title()}')
    plt.xlabel('Publisher')
    plt.ylabel(f'Sales in {region.upper()} (in millions)')
    plt.xticks(rotation=45)  # Rotate publisher names for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

 ## Sales Trend Over Time for All Regions (Including Global Sales)

```python
# Extract year from release date
df['release_year'] = df['release_date'].dt.year

# Group by release year and sum sales for each region and global sales
sales_by_year = df.groupby('release_year')[['na_sales', 'pal_sales', 'jp_sales', 'other_sales', 'total_sales']].sum()

# Plot the sales trend for all regions and global sales
plt.figure(figsize=(10, 6))
sales_by_year.plot(kind='line', marker='o', figsize=(10, 6), linewidth=2)

# Adding labels and title
plt.title('Sales Trend Over Time (By Region)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Sales (in millions)', fontsize=12)
plt.legend(title='Region', labels=['North America', 'Europe and Africa ', 'Japan', 'Other Countries', 'Global Sales'])
plt.grid(True)
plt.show()
```

## Top Performers by Average Critic Score and Total Sales Across Genre, Console, and Publisher

```python
# Grouping by genre, console, and publisher to find average critic scores and total sales
top_performers = (
    df.groupby(['genre', 'console', 'publisher'])
      .agg(
          average_critic_score=('critic_score', 'mean'),
          total_sales=('total_sales', 'sum')
      )
      .sort_values(by='average_critic_score', ascending=False)  # Sort by score
      .reset_index()
)

# Display the top 10 entries
print("Top Performers by Average Critic Score and Total Sales:")
top_performers.head(10)
```

## Top 5 Publishers by Region and Global Sales

```python
# Step 1: Define the sales columns to analyze
regions = ['na_sales', 'pal_sales', 'jp_sales', 'other_sales', 'total_sales']

# Step 2: Loop through each region to find the top 5 publishers and plot
for region in regions:
    # Group by publisher and sum sales, then find the top 5 publishers
    top_publishers = df.groupby('publisher')[region].sum().sort_values(ascending=True).head(5)
    
    # Plot the top 5 publishers for the current region
    plt.figure(figsize=(8, 5))
    top_publishers.plot(kind='barh', color='skyblue', edgecolor='black')
    plt.title(f'Top 5 Publishers by Sales in {region.replace("_sales", "").capitalize()}')
    plt.xlabel('Sales (in millions)')
    plt.ylabel('Publisher')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

## Yearly Sales Trends for Selected Genre Across Different Regions

```python
# Step 1: Select a genre to analyze
selected_genre = 'Action'  # Change this to any genre you want to analyze

# Step 2: Filter the DataFrame for the selected genre
genre_data = df[df['genre'] == selected_genre].copy()  # Use .copy() to create an independent DataFrame

# Step 3: Define sales columns and group by release year
regions = ['na_sales', 'pal_sales', 'jp_sales', 'other_sales', 'total_sales']
genre_data.loc[:, 'release_year'] = pd.to_datetime(genre_data['release_date']).dt.year  # Explicit .loc usage
yearly_sales = genre_data.groupby('release_year')[regions].sum()

# Step 4: Plot sales trends for each region
plt.figure(figsize=(10, 6))
for region in regions:
    plt.plot(yearly_sales.index, yearly_sales[region], label=region.capitalize().replace("_sales", " Sales"))

# Step 5: Customize the plot
plt.title(f'Yearly Sales Trends for {selected_genre} Genre Across Regions')
plt.xlabel('Year')
plt.ylabel('Sales (in millions)')
plt.legend()
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()
```

## Sales per Console for all regions

```python
# Define a function to plot sales per console for a specific region
def plot_console_sales(region_column, region_name):
    # Group sales by console and sum up total sales
    console_sales = df.groupby('console')[region_column].sum().sort_values(ascending=False).head(10)

    # Plot the sales per console as a horizontal bar graph
    plt.figure(figsize=(12, 6))
    console_sales.plot(kind='barh', color='lightcoral', edgecolor='black')

    # Add labels and title
    plt.title(f'Top 10 Consoles by Sales in {region_name}', fontsize=16)
    plt.xlabel('Total Sales (in million)', fontsize=12)
    plt.ylabel('Console', fontsize=12)
    plt.gca().invert_yaxis()  # To have the highest bar on top
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

# Plot for each region
plot_console_sales('na_sales', 'North America')
plot_console_sales('jp_sales', 'Japan')
plot_console_sales('pal_sales', 'PAL (Europe/Africa)')
plot_console_sales('other_sales', 'Other Regions')
plot_console_sales('total_sales', 'Global')
```
