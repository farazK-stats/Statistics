# STATISTICAL DISTRIBUTION CRITICAL VALUE CALCULATOR

```python
from scipy import stats

def stats_critical_values():
    print("STATISTICAL CRITICAL VALUE CALCULATOR\n")
    print("Choose distribution type:")
    print("1 → t-Distribution (Small sample, unknown SD)")
    print("2 → F-Distribution (ANOVA or variance comparison)")
    print("3 → Chi-Square (Independence, goodness of fit, variance test)")
    print("4 → Z-Table (Normal distribution, large sample or known SD)\n")
    
    choice = input("Enter option (1/2/3/4): ")
    
    # --------------------------- t-Distribution --------------------------- #
    if choice == "1":
        print("\n--- t-Distribution ---")
        print("Used when sample size is small (n < 30) and population SD is unknown.")
        print("Can be one-tailed or two-tailed. df = n - 1.\n")
        
        df = int(input("Enter degrees of freedom (n-1): "))
        print("Available LOS: 0.10, 0.05, 0.025, 0.01, 0.005")
        alpha = float(input("Enter level of significance (alpha): "))
        tail = int(input("Tail? 1 for one-tailed, 2 for two-tailed: "))
        
        if tail == 2:
            critical_value = stats.t.ppf(1 - alpha/2, df)
            print("\nTwo-tailed test")
            print("Critical Value =", round(critical_value, 4))
        else:
            critical_value = stats.t.ppf(1 - alpha, df)
            print("\nOne-tailed test")
            print("Critical Value =", round(critical_value, 4))
        
        t_stat = float(input("Enter your calculated t-statistic: "))
        # Interpretation
        if tail == 2:
            if abs(t_stat) > critical_value:
                print(f"Decision: |t| = {t_stat} > {round(critical_value,4)} → Reject H0")
            else:
                print(f"Decision: |t| = {t_stat} ≤ {round(critical_value,4)} → Fail to reject H0")
        else:
            if t_stat > critical_value:
                print(f"Decision: t = {t_stat} > {round(critical_value,4)} → Reject H0")
            else:
                print(f"Decision: t = {t_stat} ≤ {round(critical_value,4)} → Fail to reject H0")

    # --------------------------- F-Distribution --------------------------- #
    elif choice == "2":
        print("\n--- F-Distribution ---")
        print("Used to compare two variances or in ANOVA to test equality of means.")
        print("Right-tailed test only. df1 = numerator df, df2 = denominator df.\n")
        
        df1 = int(input("Enter numerator degrees of freedom (df1): "))
        df2 = int(input("Enter denominator degrees of freedom (df2): "))
        print("Available LOS: 0.10, 0.05, 0.01")
        alpha = float(input("Enter level of significance (alpha): "))
        
        critical_value = stats.f.ppf(1 - alpha, df1, df2)
        print("\nRight-tailed test only")
        print("Critical Value =", round(critical_value,4))
        
        F_stat = float(input("Enter your calculated F-statistic: "))
        if F_stat > critical_value:
            print(f"Decision: F = {F_stat} > {round(critical_value,4)} → Reject H0")
        else:
            print(f"Decision: F = {F_stat} ≤ {round(critical_value,4)} → Fail to reject H0")

    # --------------------------- Chi-Square --------------------------- #
    elif choice == "3":
        print("\n--- Chi-Square Distribution ---")
        print("Used for test of independence, goodness of fit, or population variance.")
        print("Right-tailed test only. df depends on test type.\n")
        
        df = int(input("Enter degrees of freedom: "))
        print("Available right-tail LOS: 0.995, 0.99, 0.975, 0.95, 0.10, 0.05, 0.025, 0.01, 0.005")
        alpha = float(input("Enter alpha value from table: "))
        
        critical_value = stats.chi2.ppf(1 - alpha, df)
        print("\nRight-tailed test only")
        print("Critical Value =", round(critical_value,4))
        
        chi_stat = float(input("Enter your calculated χ²-statistic: "))
        if chi_stat > critical_value:
            print(f"Decision: χ² = {chi_stat} > {round(critical_value,4)} → Reject H0")
        else:
            print(f"Decision: χ² = {chi_stat} ≤ {round(critical_value,4)} → Fail to reject H0")

    # --------------------------- Z-Test --------------------------- #
    elif choice == "4":
        print("\n--- Z-Table (Normal Distribution) ---")
        print("Used for large samples (n ≥ 30) or known population SD.")
        print("Can be one-tailed or two-tailed.\n")
        
        print("Available LOS: 0.10, 0.05, 0.025, 0.01, 0.005")
        alpha = float(input("Enter level of significance (alpha): "))
        tail = int(input("Tail? 1 for one-tailed, 2 for two-tailed: "))
        
        if tail == 2:
            critical_value = stats.norm.ppf(1 - alpha/2)
            print("\nTwo-tailed test")
            print("Critical Value =", round(critical_value,4))
        else:
            critical_value = stats.norm.ppf(1 - alpha)
            print("\nOne-tailed test")
            print("Critical Value =", round(critical_value,4))
        
        Z_stat = float(input("Enter your calculated Z-statistic: "))
        if tail == 2:
            if abs(Z_stat) > critical_value:
                print(f"Decision: |Z| = {Z_stat} > {round(critical_value,4)} → Reject H0")
            else:
                print(f"Decision: |Z| = {Z_stat} ≤ {round(critical_value,4)} → Fail to reject H0")
        else:
            if Z_stat > critical_value:
                print(f"Decision: Z = {Z_stat} > {round(critical_value,4)} → Reject H0")
            else:
                print(f"Decision: Z = {Z_stat} ≤ {round(critical_value,4)} → Fail to reject H0")
    
    else:
        print("Invalid option selected.")

# Run the combined program
stats_critical_values()
```
