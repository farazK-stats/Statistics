## Hypothesis Test: Comparing Median Home Values Based on Charles River Proximity

**Null Hypothesis (H₀)**: There is no significant difference in the median home values (medv)
between houses near the Charles River (chas = 1) and those not near the river (chas = 0). 

**Alternative Hypothesis (H₁)**: There is a significant difference in the median home values between the two groups.

```python
from scipy.stats import ttest_ind

# Split the data into two groups
medv_chas1 = df[df['chas'] == 1]['medv']  # Houses near the river
medv_chas0 = df[df['chas'] == 0]['medv']  # Houses away from the river

# equal_var=True (Default) → Standard T-test
# Assumes both groups have the same variance (homoscedasticity).
# Works best when sample sizes and standard deviations of the groups are similar.

# equal_var=False → Welch’s T-test
# Does NOT assume equal variance (heteroscedasticity).
# More reliable when the two groups have different variances or sample sizes.

# Perform an independent t-test
t_stat, p_value = ttest_ind(medv_chas1, medv_chas0, equal_var=False)  # Welch’s T-test
```

```python
# Print results
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

**Conclusion:** 
 Since p-value < 0.05, we reject the null hypothesis (H₀) and conclude that proximity to the Charles River has an impact on housing prices.
 There is a significant difference in the median home values between houses near the Charles River and those away from it (p-value = 0.0035)

## Hypothesis Test: Comparing Median Home Values (medv) Across House Age Groups (age)

**Null Hypothesis (H₀):** There is no significant difference in median home values (medv) among different age groups of houses.  
**Alternative Hypothesis (H₁):** There is a significant difference in median home values between at least one of the age groups.

```python
from scipy.stats import f_oneway

# Group the age Variable into Categories
# Create age groups
young = df[df['age'] <= 35]['medv']
middle_aged = df[(df['age'] > 35) & (df['age'] <= 70)]['medv']
old = df[df['age'] > 70]['medv']

# Perform ANOVA test
f_stat, p_value = f_oneway(young, middle_aged, old)

# Print results
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

**Conclusion:**
Since p-value (1.71e-15) < 0.05, we reject the null hypothesis (H₀).This means that at least one of the age groups
(young, middle-aged, or old houses) has a significantly different median home value compared to the others.

## Hypothesis Test: Relationship Between Nitric Oxide (nox) and Non-Retail Business Acres (indus)

**Null Hypothesis (H₀):** There is no significant correlation between Nitric Oxide concentration (nox) and the proportion of non-retail business acres (indus).      
**Alternative Hypothesis (H₁):** There is a significant correlation between nox and indus.

```python
from scipy.stats import pearsonr

# Compute Pearson correlation
corr_coef, p_value = pearsonr(df['nox'], df['indus'])

# Print results
print("Pearson Correlation Coefficient:", corr_coef)
print("P-value:", p_value)
```

**Conclusion:**
There is a strong positive correlation between the Nitric Oxide concentration (nox) and the proportion of non-retail business acres (indus) with a Pearson correlation coefficient of 0.76. This suggests that towns with a higher proportion of industrial land tend to have higher levels of Nitric Oxide pollution. The p-value of 7.91e-98 is much smaller than 0.05, which allows us to reject the null hypothesis (H₀) and conclude that there is a significant relationship between these two variables. Industrial areas are likely a major contributor to increased air pollution, particularly in terms of Nitric Oxide emissions.

## Hypothesis Test: Impact of Weighted Distance to Employment Centers on Median Home Value

**Null Hypothesis (H₀):** There is no significant relationship between the weighted distance to employment centers (dis) and the median value of owner-occupied homes (medv).       
**Alternative Hypothesis (H₁):** There is a significant relationship between the weighted distance to employment centers (dis) and the median value of owner-occupied homes (medv).

```python
# We will use linear regression to analyze the relationship between dis (distance) and medv (median home value).

import statsmodels.api as sm

# Define the independent variable (X) and dependent variable (y)
X = df['dis']  # weighted distance to employment centers
y = df['medv']  # median value of owner-occupied homes

# Add a constant to the independent variable (for the intercept in the regression equation)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```

**Conclusion:**   
The regression analysis shows a positive relationship between the distance to employment centers (dis) and the median value of homes (medv). The regression equation is: **medv = 18.39+1.09*DIS**. This means that for each additional unit of distance from employment centers, the median home value increases by $1,091.60. The p-value for dis is 0.000, which is less than 0.05, indicating a statistically significant relationship. Thus, we conclude that higher distances to employment centers are associated with higher median home values.
